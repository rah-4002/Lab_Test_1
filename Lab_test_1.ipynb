```python
City_A =np.array([30, 32, 31, 29, 28, 27, 26])
City_B =np.array([35, 34, 40, 33, 32, 31, 30])
City_C =np.array([25, 26, 27, 28, 29, 30, 31])
City_D =np.array([22, 23, 24, 25, 26, 27, 36])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def avg_min_max(arr):
    avg_temp=np.mean(arr)
    print("Average Temp: ",avg_temp)
    min_temp=np.min(arr)
    print("Minimum Temp: ",min_temp)
    max_temp=np.max(arr)
    print("Maximum Temp: ",max_temp)
```

```python
avg_min_max(City_A)
```

```python
avg_min_max(City_B)
```

```python
avg_min_max(City_C)
```

```python
avg_min_max(City_D)
```

```python
def avg_per_day(a1,a2,a3,a4):
    for i in range(7):
        avg=(a1[i]+a2[i]+a3[i]+a4[i])/4
        print("Average of ",i+1," day is: ",avg)
```

```python
avg_per_day(City_A,City_B,City_C,City_D)
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.arange(1,8),City_A,color="blue")
plt.plot(np.arange(1,8),City_B,color="green")
plt.plot(np.arange(1,8),City_C,color="purple")
plt.plot(np.arange(1,8),City_D,color="red")
plt.title("Temperature v/s Days")
plt.xlabel("Days of Week")
plt.ylabel("Temperature")
plt.legend("best_loc")
plt.grid()
plt.show()
```
